```python
import yfinance as yf
import pyfolio as pf
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# I Invested in EFT, John Hancock Multifactor Technology (JHMT)
jhmt, dgifx = yf.Ticker('JHMT'), yf.Ticker('DGIFX')
history = jhmt.history('max'), dgifx.history('max')
for i in range(len(history)):
    history[i].index = history[i].index.tz_localize('utc')
    print(history[i].info())
```

```python
returns = [history[i].Close.pct_change() for i in range(len(history))]
print(returns)
```

```python
i = 0 # JHMT
pf.create_returns_tear_sheet(returns[i], live_start_date='2020-1-1')
```
